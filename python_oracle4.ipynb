```python
#fas-정재환
import cx_Oracle
import pandas as pd
```

```python
#오라클 서버  접속
conn = cx_Oracle.connect('scott/c4iuser11@localhost:1521/orcl')
#커서 연결
cursor = conn.cursor()
```

```python
# 행사 제목의 첫글자가 e 보다 크고 r 보다 작은 행사를 모두 출력
sql  = "select * from seminar where title > 'e%' and title < 'r%' " 
sql += "order by num"
rs = pd.read_sql(sql, conn)
rs
```

```python
# 지역별 행사 개최 횟수 출력
sql  = "select area, count(*) from seminar group by area" 
rs = pd.read_sql(sql, conn)
rs
```

```python
# 분야별 행사 개최 횟수 출력
sql  = "select field, count(*) from seminar group by field " 
sql += "order by count(*) desc"
rs = pd.read_sql(sql, conn)
rs
```

```python
# 제목에 '기'이 포함된 행사들의 제목을 앞에서 3글자만 출력
sql  = "select substr(title, 1, 3) from seminar where title like '%기%' " 
sql += "order by title"
rs = pd.read_sql(sql, conn)
rs
```

```python
# 제목이 40글자 이상이 행사 정보를 출력
sql  = "select * from seminar where length(title) >= 40 " 
sql += "order by num"
rs = pd.read_sql(sql, conn)
rs
```

```python
# 제목이 40글자 이상이 행사 정보를 출력
sql  = "select field, avg(length(title)) from seminar group by field " 
sql += "order by field"
rs = pd.read_sql(sql, conn)
rs
```
