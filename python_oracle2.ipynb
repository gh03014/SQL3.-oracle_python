```python
# 테이블에서 첫번째 정렬은 신장, 두번째 정렬을 체중, 세번째 정렬은 혈당으로 내림차순한다
sql = 'select 신장, 체중, 식전혈당 from health order by 신장, 체중, 식전혈당 desc'
rs = pd.read_sql(sql, conn)
rs
```

```python
# 체중이 95인 사람의 수축혈압을 20으로 나눈 나머지를 구하여라
sql = 'select 번호, 체중, round(수축기혈압/20) from health where 체중 = 95'
rs = pd.read_sql(sql, conn)
rs
```

```python
# 번호가 150~333인 사람의 체중에 10000000을 곱하고 세자리 마다 ','를 출력하고 맨앞에 '$'를 출력한다
sql = "select 번호, to_char(체중*10000000, '$999,999,999') from health where 번호 >= 150 and 번호 < 334"
rs = pd.read_sql(sql, conn)
rs
```

```python
# 체중을 기준으로 그룹화 한 후 인원이 20000명 이상인 그룹의 체중, 인원수를 구한다
sql = "select 체중, count(*) from health group by 체중 having count(*) > 20000"
rs = pd.read_sql(sql, conn)
rs
```
